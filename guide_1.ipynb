### So, Let's Get Started - 
When it comes to simplicity, at an abstract level you can directly use matplotlib by just importing pyplot as plt and using the plot function.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,10,15)
y = x**2

plt.plot(x,y)
```

But using the object oriented approach is more intuitive way and it becomes convenient when handling multiple axes and plots together.<br>
 - First things first, Let create a figure object. This represents an empty canvas where you can add a set of axes to plot something.<br>
 - Now, we can add a set of axes to the figure using add_axes() method. This takes a list as argument which contains four values - __left__, __bottom__, __width__, __height__ i.e. the position of axes inside the canvas. All these values lies between 0 and 1.
 - And Finally, you can plot on the axes using plot() method.

```python
fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x,y)
```

This may seem harder than the typical way, but what if I say that you don't have to create figure and add axes to it by your own. Yes, this can be done using the __subplots()__ function. The __plt.subplots()__ returns a tuple with a figure object and an array of axes ( bydefault single axes) which can be used to plot directly. Let's make the plot a little more attractive and readable by - 
 - Adding a grid with grid() method.
 - 'r--' parameter in plot() method is for dashed red line.
 - lw=2 sets line width as double of original width.
 - marker='o', markeredgecolor='blue', markersize=10 denotes marker shape as round with blue edge color and size 10.
 - Specifying title and labels for the plot.

```python
fig, ax = plt.subplots()
ax.grid() # for grid
ax.plot(x, y, 'r--', lw=2, marker='o', markeredgecolor='blue', markersize=10)
ax.set_title('My Data Plot')
ax.set_xlabel('X label Here')
ax.set_ylabel('Y label Here')
```

A single set of axes can also be used to plot multiple items. These plots can be separated by adding labels to each of them. These labels are displayed as legend when you call the __ax.legend()__ method. You can also add location to the legend by adding the loc parameter to legend. By default, it is set to the best found location.

```python
fig, ax = plt.subplots()

ax.grid()
ax.plot(x, x**2, 'b--', label='^2')
ax.plot(x, x**2.5, 'r', marker='^', label='^2.5')
ax.plot(x, x**3, 'purple', marker='*', label='^3')
ax.legend()

ax.set_title('My Data Plot')
ax.set_xlabel('X label Here')
ax.set_ylabel('Y label Here')
```

As I mentioned earlier that __plt.subplots()__ returns a tuple with figure object and an array of axes. By default, the parameters _nrows_ and _ncols_ are set to 1 that's why we were getting a single axes object. Now, changing the value of _nrows_ and _ncols_ to 2 we get _ax_ as an array of Axes with 4 objects.<br><br>
_NOTE :-_ You can also change the figure size as the requirements by passing a tuple to _figsize_ parameter. This tuple denotes the _length_ and _width_ of the figure in inches. Here, I took _figsize=(12,8)_.

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
```

 
Having a look at the shape, type and content of _ax_, we can clearly see that it is a 2x2 numpy's ndarray.

```python
print(ax.shape)
print('---------------------------------------')
print(type(ax))
print('---------------------------------------')
print(ax)
```

These AxesSubplot objects can be accessed simply by indices or by iterating over the array.

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

a = np.random.randn(500)
ax[0][0].hist(a, bins= 50, color='purple')
ax[0][0].set_title('Gaussian Distributed Plot')

a = np.arange(50)
b = np.sin(a)
ax[0][1].grid()
ax[0][1].plot(a,b,'b', marker='o')
ax[0][1].set_title('My Sine Plot')

ax[1][0].grid()
ax[1][0].plot(x, x**2, 'b--', label='^2')
ax[1][0].plot(x, x**2.5, 'r', marker='^', label='^2.5')
ax[1][0].plot(x, x**3, 'purple', marker='*', label='^3')
ax[1][0].legend()

ax[1][1].grid()
ax[1][1].plot(x,y)
ax[1][1].set_title('My Data Plot')
ax[1][1].set_xlabel('X label Here')
ax[1][1].yaxis.set_label_position("right")
ax[1][1].set_ylabel('Y label Here')

fig.tight_layout()
```

_NOTE :-_ _fig.tight__layout()_ is used to ensure that two AxesSubplots do not overlap on each other.
